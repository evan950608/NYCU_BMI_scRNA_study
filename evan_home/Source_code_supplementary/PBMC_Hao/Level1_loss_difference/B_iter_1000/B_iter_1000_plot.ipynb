```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
```

```python
with open('B_loss_history.json', 'r') as f:
    B_loss_history = json.load(f)
with open('B_loss_diff_history.json', 'r') as f:
    B_loss_diff_history = json.load(f)
```

```python
iteration = np.arange(0, 1000, 1)

# Set figure size
plt.figure(figsize=(6, 3))

# Plot loss history
plt.plot(iteration, B_loss_history.values())

# Plot red vertical line at x=477
plt.axvline(x=477, color='red')

# Set title and labels
plt.title('B cells loss history')
plt.xlabel('Iteration')
plt.ylabel('Loss')

plt.show()
```

```python
iteration = np.arange(0, 1000, 1)

# Set figure size
plt.figure(figsize=(6, 3))

# Plot loss history
plt.plot(iteration, B_loss_diff_history.values())

# Plot red vertical line at x=477
plt.axvline(x=477, color='red')
plt.axhline(y=1e-6, color='red', linestyle='--')

# Set x-axis limit
# plt.ylim(top = 0.1)
plt.yscale('log')

# Set title and labels
plt.title('B cells loss difference history')
plt.xlabel('Iteration')
plt.ylabel('Loss difference')

plt.show()
```
