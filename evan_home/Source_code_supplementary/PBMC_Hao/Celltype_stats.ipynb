```python
data = sc.read_h5ad('/Users/evanli/Documents/Research_datasets/PBMC_Hao/Hao_PBMC.h5ad')
data.shape
```

```python
data.obs['celltype.l1'] = data.obs['celltype.l1'].str.replace(' ', '_')
data.obs['celltype.l2'] = data.obs['celltype.l2'].str.replace(' ', '_')
l1_types = np.unique(data.obs['celltype.l1']).tolist()
l2_types = np.unique(data.obs['celltype.l2']).tolist()
```

```python
# Plot Level 1
counts = data.obs['celltype.l1'].value_counts()
counts = counts.sort_index()
counts.plot(kind='bar')
plt.title('Level_1 cell type counts')
plt.ylabel('Counts')
plt.show()
```

```python
# Plot Level 2
counts = data.obs['celltype.l2'].value_counts()
counts = counts.sort_index()
counts.plot(kind='bar')
plt.title('Level_2 cell type counts')
plt.ylabel('Counts')
plt.show()
```
