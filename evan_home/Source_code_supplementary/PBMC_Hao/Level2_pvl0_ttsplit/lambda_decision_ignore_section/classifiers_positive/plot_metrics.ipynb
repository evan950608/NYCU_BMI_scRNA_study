```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
regular_all_metrics = pd.read_csv('L2_regular_Kfold_metrics.csv', index_col=0)
regular_all_metrics.head()
```

```python
stratified_all_metrics = pd.read_csv('L2_stratified_Kfold_metrics.csv', index_col=0)
stratified_all_metrics.head()
```

```python
# Plot Regular K-fold metrics for each celltype
ax = regular_all_metrics.plot.bar(rot=0, figsize=(15, 6), title='One vs. Rest Logistic Regression 5-fold CV (positive features)')
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
# Rotate x-ticks by 90 degrees
plt.xticks(rotation=90)
plt.savefig('L2_positive_classifier_Kfold_metrics.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
# Plot Stratified K-fold metrics for each celltype
ax = stratified_all_metrics.plot.bar(rot=0, figsize=(15, 6), title='One vs. Rest Logistic Regression stratified 5-fold CV (positive features)')
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
# Rotate x-ticks by 90 degrees
plt.xticks(rotation=90)
plt.savefig('L2_positive_classifier_strat_Kfold_metrics.png', dpi=300, bbox_inches='tight')
plt.show()

```
