```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
annot = pd.read_csv("/Users/evanli/Documents/Research_datasets/PBMC_68k_Zheng/github_data/68k_pbmc_barcodes_annotation.tsv", sep="\t", index_col=2)
annot.head()
```

```python
annot['celltype'].unique()
```

```python
plt.figure(figsize=(10, 8))
for celltype in annot['celltype'].unique():
    plt.scatter(annot[annot['celltype'] == celltype]['TSNE.1'], 
                annot[annot['celltype'] == celltype]['TSNE.2'], 
                label=celltype, s=2, alpha=0.5)  # Make dots smaller
leg = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend to the right of plot
for handle in leg.legendHandles:
    handle._sizes = [30]  # Make legend dots bigger
    handle.set_alpha(1)  # Make legend dots solid

plt.xlabel('TSNE.1')
plt.ylabel('TSNE.2')
plt.title('tsne.1 vs tsne.2 colored by celltype')
plt.show()
```
