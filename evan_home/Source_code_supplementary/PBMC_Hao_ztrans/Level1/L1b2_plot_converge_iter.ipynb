```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import os
```

```python
os.chdir(r"C:\Users\evanlee\Documents\GitHub\EvanPys\Progress\PBMC_Hao_ztrans\Level1\tuning_result")

# ['B', 'CD4_T', 'CD8_T', 'DC', 'Mono', 'NK', 'other', 'other_T']
celltype = 'B'
with open('{}_loss_history.json'.format(celltype), 'r') as f:
    loss_history_dict = json.load(f)
```

```python
loss_history_dict.keys()
```

```python
loss_history_dict['0']
```

```python
for key in loss_history_dict.keys():
    loss_history = loss_history_dict[key]
    iteration = list(int(i) for i in loss_history.keys())
    loss = list(loss_history.values())
    plt.plot(iteration, loss)
    plt.title('Lambda_{} loss history'.format(key))
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.show()
    
```

```python
celltype = 'B'
with open('{}_loss_diff_history.json'.format(celltype), 'r') as f:
    loss_diff_history_dict = json.load(f)
```

```python
for key in loss_diff_history_dict.keys():
    loss_diff = loss_diff_history_dict[key]
    iteration = list(int(i) for i in loss_diff.keys())
    loss = list(loss_diff.values())
    plt.plot(iteration, loss)
    plt.title('Lambda_{} loss history'.format(key))
    plt.yscale('log')
    plt.xlabel('Iteration')
    plt.ylabel('Loss difference')
    plt.show()
    
```
