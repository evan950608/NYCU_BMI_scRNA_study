```python
# full Stuart bone marrow dataset
# adata = sc.read_h5ad('/home/jovyan/work/Research_datasets/Stuart_bm/Stuart_bm.h5ad')
adata = sc.read_h5ad(r"C:\Users\evanlee\Documents\Research_datasets\Stuart_bm\Stuart_bm_v2.h5ad")
print('Original adata:', adata.shape)
```
