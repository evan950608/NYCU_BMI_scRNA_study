```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
```

```python
with open('CD4_T_loss_history.json', 'r') as f:
    cd4t_loss_history = json.load(f)
with open('CD4_T_loss_diff_history.json', 'r') as f:
    loss_diff_history = json.load(f)
```

```python
stop_iter = 560
```

```python
cd4t_loss_history
```

```python
# modify first iter
# cd4t_loss_history['0'] = cd4t_loss_history['1']

iteration = np.arange(2, 1000, 1)

# Set figure size
plt.figure(figsize=(6, 3))

# Plot loss history
plt.plot(iteration, list(cd4t_loss_history.values())[2:])

# Plot red vertical line at x=477
plt.axvline(x=stop_iter, color='red')

# Set title and labels
plt.title('CD4 T cells loss history')
plt.xlabel('Iteration')
plt.ylabel('Loss')

plt.show()
```

```python
loss_diff_history
```

```python
# modify first two iters
# loss_diff_history['1'] = loss_diff_history['2']

iteration = np.arange(2, 1000, 1)

# Set figure size
plt.figure(figsize=(6, 3))

# Plot loss history
plt.plot(iteration, list(loss_diff_history.values())[2:])

# Plot red vertical line at x=477
plt.axvline(x=stop_iter, color='red')
plt.axhline(y=1e-6, color='red', linestyle='--')

# Set x-axis limit
# plt.ylim(top = 0.1)
plt.yscale('log')

# Set title and labels
plt.title('CD4 T cells loss difference history')
plt.xlabel('Iteration')
plt.ylabel('Weight difference')

plt.show()
```
