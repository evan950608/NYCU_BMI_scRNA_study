```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
performance_df = pd.read_csv('Stuart_annot_performance.csv', index_col=0)
performance_df
```

```python
# plot barplot
# Clean up column names to remove leading/trailing spaces
performance_df.columns = performance_df.columns.str.strip()

# Retry filtering the data with cleaned column names
l1_performance = performance_df.loc[['PreLect LR L1 (Hao)', 'SingleR L1 (Hao)', 'CellAssign L1'], ]
l1_performance.index = l1_performance.index.str.replace(' (Hao)', '')
l2_performance = performance_df.loc[['PreLect LR L2', 'SingleR L2', 'CellAssign L2'], ]

```

```python
l1_performance
```

```python
l2_performance
```

```python
# plot L1
ax = l1_performance.plot.bar(title='Performance comparison of Stuart Level 1 classification')
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# plot L1
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
l1_performance.plot.bar(ax=axes[0], title='Performance comparison of Stuart Level 1 classification')
l2_performance.plot.bar(ax=axes[1], title='Performance comparison of Stuart Level 2 classification')
axes[0].get_legend().remove()
axes[0].set_ylabel('Metrics')
axes[1].legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
axes[1].set_ylabel('Metrics')
plt.tight_layout()
plt.show()
```
